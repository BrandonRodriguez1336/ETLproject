```python
import requests
import lxml.html as lh
import pandas as pd
```

```python
url='https://www.payscale.com/data/gender-pay-gap'
```

```python
tables = pd.read_html(url)
html = tables[0]
```

```python

html = html.copy()
```

```python
new_header = html.iloc[0] #grab the first row for the header
html = html[1:] #take the data less the header row
html.columns = new_header #set the header row as the df header
```

```python
#clean up the dollar signs and percent signs
html['Uncontrolled Gender Pay Gap'] = html['Uncontrolled Gender Pay Gap'].replace({'\$':''}, regex = True)
html['Controlled Gender Pay Gap'] = html['Controlled Gender Pay Gap'].replace({'\$':''}, regex = True)
html['Percent Men (BLS)'] = html['Percent Men (BLS)'].replace({'\%':''}, regex = True)
html['Percent Women (BLS)'] = html['Percent Women (BLS)'].replace({'\%':''}, regex = True)
html
```
